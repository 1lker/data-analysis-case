# A/B Test Analysis of Monetization Strategy

## Dataset Overview
We are analyzing data from a monetization A/B test where users joined between **October 2, 2022**, and **October 22, 2022**. The dataset includes data until **December 30, 2022**. Our goal is to evaluate which variant is more effective at maximizing users' **Lifetime Value (LTV)** over a 365-day period (**d365**). We will also consider an additional scenario where the strategy changes to a 366-day period (**d366**).

## Analysis Questions

### a) Which variant is winning in LTV? Why? How would you interpret the difference between the variants?

#### Steps for Analysis:
1. **Calculate LTV for both variants**:
   - Define LTV as the **total revenue per user** for each variant, considering only the time period up to December 30, 2022.
   - Since the data is available for only a few months after user acquisition, we'll estimate the total LTV using the observed data and project the revenue forward.

2. **Compare the Variants**:
   - Compute the **average LTV** for users in each variant.
   - Use appropriate statistical tests (e.g., t-tests or Mann-Whitney U tests) to determine if the difference in LTV between the variants is statistically significant.

#### Interpretation:
- The **winning variant** is the one with the higher average LTV.
- **Interpretation of the Difference**:
   - A significant difference in LTV between the two variants could indicate that one monetization strategy leads to more effective user engagement and revenue generation over time.
   - Factors such as **user behavior**, **purchase patterns**, or the effectiveness of incentives in the variant could explain the observed differences.
   - If the difference is not statistically significant, it might suggest that the variants are similar in effectiveness, or that more time/data is needed to draw conclusions.

---

### b) (Bonus) What would be your additional comments, if we change our strategy to d366 from d365?

#### Considerations:
- **Incremental Impact**:
   - Shifting the analysis from **d365 to d366** would only introduce a **1-day extension** in the user lifetime.
   - Given that revenue typically follows a **decay curve** over time, we expect the additional day to contribute **marginal revenue**.

#### Potential Insights:
- The **effect on LTV** will likely be **minimal**, as most monetization occurs earlier in a user’s lifetime.
- However, if there are specific **timing-based incentives or promotions** active on that extra day (e.g., special events or sales), it could slightly boost the revenue on **d366**.
- Overall, changing from **d365 to d366** would not drastically alter our conclusions but could provide marginal incremental revenue, depending on the specific user behavior on that day.

2. Hypothesis Testing
Null Hypothesis (H₀): There is no difference in the average LTV between users in Variant A and Variant B.
Alternative Hypothesis (H₁): There is a significant difference in the average LTV between users in Variant A and Variant B.
We will calculate the Lifetime Value (LTV) by summing all the revenue associated with each user up until December 30, 2022, and then compare the LTVs between the two experiment variants.


3. LTV (Lifetime Value) Hesaplama


```python
# Import bootstrapping functionality for more robust confidence intervals
!pip install scikit-learn
from sklearn.utils import resample
import numpy as np

# Assuming ltv is the DataFrame with user_id, experiment_variant, and revenue columns
# Calculate LTV from the original cleaned data
ltv = data_cleaned.groupby(['user_id', 'experiment_variant'])['revenue'].sum().reset_index()

# Define ltv_A and ltv_B
ltv_A = ltv[ltv['experiment_variant'] == 'A']['revenue']
ltv_B = ltv[ltv['experiment_variant'] == 'B']['revenue']

# Check that ltv_A and ltv_B are correctly defined
print("First 5 records of ltv_A:\n", ltv_A.head())
print("First 5 records of ltv_B:\n", ltv_B.head())


# Bootstrapping for average LTV estimates
def bootstrap_mean(data, n_iterations=1000):
    boot_means = []
    for _ in range(n_iterations):
        sample = resample(data)
        boot_means.append(sample.mean())
    return np.percentile(boot_means, [2.5, 97.5])

# Get the LTV estimates with bootstrapped confidence intervals
ltv_boot_A = bootstrap_mean(ltv_A)
ltv_boot_B = bootstrap_mean(ltv_B)

print(f"95% Confidence Interval for Variant A LTV: {ltv_boot_A}")
print(f"95% Confidence Interval for Variant B LTV: {ltv_boot_B}")
```

```python
import matplotlib.pyplot as plt

# Plot confidence intervals
variants = ['Variant A', 'Variant B']
means = [ltv_A.mean(), ltv_B.mean()]
ci_lower = [ltv_boot_A[0], ltv_boot_B[0]]
ci_upper = [ltv_boot_A[1], ltv_boot_B[1]]

plt.figure(figsize=(8, 6))
plt.errorbar(variants, means, yerr=[ci_upper - means, means - ci_lower], fmt='o', capsize=5, capthick=2, color='blue', ecolor='black')
plt.title('95% Confidence Intervals for LTV of Variants A and B')
plt.ylabel('Lifetime Value (LTV)')
plt.show()
```

```python
# Kullanıcı başına toplam gelir (LTV) hesapla
ltv = data_cleaned.groupby(['user_id', 'experiment_variant'])['revenue'].sum().reset_index()

# İlk ve son 5 satırı göster
print("LTV hesaplamasının ilk 5 satırı:\n", ltv.head())
print("LTV hesaplamasının son 5 satırı:\n", ltv.tail())

# Varyant bazında ortalama LTV hesapla
ltv_per_variant = ltv.groupby('experiment_variant')['revenue'].mean().reset_index()
ltv_per_variant.columns = ['experiment_variant', 'average_ltv']

# Ortalama LTV değerlerini göster
print("Varyant bazında ortalama LTV:\n", ltv_per_variant)
```

# 4. Welch’s t-Test ile İstatistiksel Test

```python
from scipy.stats import ttest_ind

# LTV değerlerini varyanta göre ayır
ltv_A = ltv[ltv['experiment_variant'] == 'A']['revenue']
ltv_B = ltv[ltv['experiment_variant'] == 'B']['revenue']

# Welch's t-Test uygula
t_stat, p_value = ttest_ind(ltv_A, ltv_B, equal_var=False)

# Sonuçları göster
print(f"T-istatistiği: {t_stat}, P-değeri: {p_value}")
```

# 5. Cohen's d ile Etki Büyüklüğü

```python
import numpy as np

# Cohen's d hesapla
mean_diff = np.mean(ltv_A) - np.mean(ltv_B)
pooled_std = np.sqrt((np.var(ltv_A, ddof=1) + np.var(ltv_B, ddof=1)) / 2)
cohens_d = mean_diff / pooled_std

# Cohen's d sonucunu göster
print(f"Cohen's d: {cohens_d}")
```

### Cohen's d Interpretation:
- 0.2: Small effect
- 0.5: Medium effect
- 0.8: Large effect This gives us more context on the practical significance of the results.

```python
# Complete A/B Test Analysis Notebook

# Import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import resample
from scipy.stats import ttest_ind
import warnings

# Suppress warnings for clean output
warnings.filterwarnings("ignore")

# Function to load and clean the dataset
def load_and_clean_data(file_path):
    # Load data
    data = pd.read_json(file_path, lines=True)

    # Convert timestamps to datetime
    data['event_time'] = pd.to_datetime(data['event_time'], unit='ms')
    data['first_event_time'] = pd.to_datetime(data['first_event_time'], unit='ms')

    # Extract the relevant properties
    data['revenue'] = data['properties'].apply(lambda x: next((item['value'].get('float_value') for item in x if item['key'] == 'revenue'), 0))
    data['product_duration'] = data['properties'].apply(lambda x: next(item['value'].get('string_value') for item in x if item['key'] == 'productDuration'))

    # Select important columns
    data_cleaned = data[['experiment_variant', 'user_id', 'event_name', 'first_event_time', 'event_time', 'revenue']]

    return data_cleaned

# Function to calculate LTV per variant
def calculate_ltv(data_cleaned):
    # Group by user and experiment variant, and calculate total revenue (LTV)
    ltv = data_cleaned.groupby(['user_id', 'experiment_variant'])['revenue'].sum().reset_index()
    return ltv

# Bootstrapping function for confidence intervals
def bootstrap_mean(data, n_iterations=1000):
    boot_means = []
    for _ in range(n_iterations):
        sample = resample(data)
        boot_means.append(sample.mean())
    return np.percentile(boot_means, [2.5, 97.5])

# Visualization of distributions and confidence intervals
def visualize_distributions(ltv_A, ltv_B, ltv_boot_A, ltv_boot_B):
    plt.figure(figsize=(12, 6))
    
    # KDE plot for distributions
    sns.kdeplot(ltv_A, label='Variant A', shade=True, color="blue")
    sns.kdeplot(ltv_B, label='Variant B', shade=True, color="orange")
    plt.title('Revenue Distribution for Variants A and B')
    plt.xlabel('Revenue')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    # Plot confidence intervals
    variants = ['Variant A', 'Variant B']
    means = [ltv_A.mean(), ltv_B.mean()]
    ci_lower = [ltv_boot_A[0], ltv_boot_B[0]]
    ci_upper = [ltv_boot_A[1], ltv_boot_B[1]]

    # Convert lists to numpy arrays for arithmetic operations
    means = np.array(means)
    ci_lower = np.array(ci_lower)
    ci_upper = np.array(ci_upper)

    plt.figure(figsize=(8, 6))
    plt.errorbar(variants, means, yerr=[means - ci_lower, ci_upper - means], fmt='o', capsize=5, capthick=2, color='blue', ecolor='black')
    plt.title('95% Confidence Intervals for LTV of Variants A and B')
    plt.ylabel('Lifetime Value (LTV)')
    plt.show()


# Function to perform Welch's t-test and Cohen's d for effect size
def perform_statistical_tests(ltv_A, ltv_B):
    # Welch's t-test
    t_stat, p_value = ttest_ind(ltv_A, ltv_B, equal_var=False)
    
    # Cohen's d for effect size
    mean_diff = np.mean(ltv_A) - np.mean(ltv_B)
    pooled_std = np.sqrt((np.var(ltv_A, ddof=1) + np.var(ltv_B, ddof=1)) / 2)
    cohens_d = mean_diff / pooled_std

    return t_stat, p_value, cohens_d

# Final results display with summary
def display_results(t_stat, p_value, cohens_d, ltv_boot_A, ltv_boot_B):
    print("### A/B Test Results Summary ###")
    print(f"Welch's t-test: T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    
    # Interpret the results of the p-value
    if p_value < 0.05:
        print("Conclusion: The difference in LTV between Variant A and Variant B is statistically significant (p < 0.05).")
    else:
        print("Conclusion: The difference in LTV between Variant A and Variant B is not statistically significant (p >= 0.05).")
    
    print(f"Cohen's d (Effect Size): {cohens_d:.4f}")
    if abs(cohens_d) < 0.2:
        print("Effect size interpretation: Negligible")
    elif abs(cohens_d) < 0.5:
        print("Effect size interpretation: Small")
    elif abs(cohens_d) < 0.8:
        print("Effect size interpretation: Medium")
    else:
        print("Effect size interpretation: Large")
    
    # Display bootstrapped confidence intervals
    print(f"95% Confidence Interval for Variant A LTV: {ltv_boot_A}")
    print(f"95% Confidence Interval for Variant B LTV: {ltv_boot_B}")

# Main function to execute the analysis
def main():
    # Load and clean data
    file_path = 'data/dataset_experiment.json'  # Replace with your file path
    data_cleaned = load_and_clean_data(file_path)
    
    # Calculate LTV
    ltv = calculate_ltv(data_cleaned)
    
    # Split LTV by variant
    ltv_A = ltv[ltv['experiment_variant'] == 'A']['revenue']
    ltv_B = ltv[ltv['experiment_variant'] == 'B']['revenue']
    
    # Bootstrapping for confidence intervals
    ltv_boot_A = bootstrap_mean(ltv_A)
    ltv_boot_B = bootstrap_mean(ltv_B)

    # Visualize distributions and confidence intervals
    visualize_distributions(ltv_A, ltv_B, ltv_boot_A, ltv_boot_B)

    # Perform statistical tests
    t_stat, p_value, cohens_d = perform_statistical_tests(ltv_A, ltv_B)

    # Display final results and summary
    display_results(t_stat, p_value, cohens_d, ltv_boot_A, ltv_boot_B)

# Run the analysis
main()
```
